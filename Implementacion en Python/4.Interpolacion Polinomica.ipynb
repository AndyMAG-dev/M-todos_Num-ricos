Cuando te dicen interpolacion de grado TAL, se necesitan (TAL + 1) NODOS

Lineal significa X , Cuadratico significa X**2 (se necesita 1 nodo mas) 

### Consideraciones:
- las 2 primeras X siempre deben encerrar al punto que estas buscando
- Luego el resto de X elegir las mas cercanas

```python
import numpy as np 
from sympy import *


Equiss = np.array([-1.0 ,0.0, 1.0])
Yess = np.array([0.37, 1.0, 2.72])       # Solo SE modifican estos 3
X = 0.5

cantNodos = Equiss.size

x = symbols('x')
L = {}
Lsust = {}

i = 0
j = 0
numerador = 1
denominador = 1


while i < cantNodos:
    while j < cantNodos:
        if j != i:
            numerador *= (x - Equiss[j])
            denominador *= (Equiss[i] - Equiss[j])
        j += 1
    L[i] = numerador / denominador
    i+=1
    j = 0
    numerador = 1
    denominador = 1
    
    
while j < cantNodos:
    print(f"L({j}) = {L[j]}")
    print(f"L({j}) = {expand(L[j])}")
    print("")
    j += 1
    
j=0
p=0
while j < cantNodos:
    p = p + (Yess[j] * L[j])
    j +=1

print(f"P(X) = {expand(p)}")

Px = lambdify(x,p)

print(f"P({X}) = {Px(X)} ")
```

### ERROR DE LAGRANGE
- El error solo necesita que escribas la funcion, no tienes que escribir
mas nada xq solo necesita tus datos de X que tienes arriba
- La respuesta la da en negativo o positivo da igual

```python
def mayorX():
    i = 0
    mayor = -99999999999
    pos = -1
    while i < cantNodos:
        evaluacion = abs(df3funcion(Equiss[i]))   ###poner aqui la derivada que sea necesaroa
        if mayor < evaluacion:
            mayor = evaluacion
            pos = i
        i+=1
    return Equiss[pos]


f = "E**x"                           ### modificar aqui la funcion
df = diff(f)                       ### Derivar tantos nodos tenga
df2 = diff(df)
df3 = diff(df2)
df3funcion = lambdify(x,df3)    ### hacerle el lambdify a la derivada correspondiente
c = mayorX()

print(f"C = {c}")
print(f"derivaca evaluada en C = {df3funcion(c)}")          ### poner la derivada correspondiente
print(f"n = {cantNodos}")
print(f"n! = {factorial(cantNodos)}")

Error = df3funcion(c) / factorial(cantNodos)                     ### poner la derivada correspondiente

i = 0
while i < cantNodos:
    Error *= (X - Equiss[i])
    
    i+=1

print(f"derivada correspondiente: {df3}")
print(f"Error = {Error}")
print("Error = {:.10f}".format(Error))

        
```

### Newton
Los 2 primeros puntos son los que contienen a la X, y el resto se hace por cercania

```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt

# INGRESO , Datos de prueba
xi = np.array([14.0 , 16.0, 18.0])
fi = np.array([6.4 , 5.9, 5.9])
Xvalor = 15
# PROCEDIMIENTO

# Tabla de Diferencias Divididas Avanzadas
titulo = ['i','xi','fi']
n = len(xi)
ki = np.arange(0,n,1)
tabla = np.concatenate(([ki],[xi],[fi]),axis=0)
tabla = np.transpose(tabla)

# diferencias divididas vacia
dfinita = np.zeros(shape=(n,n),dtype=float)
tabla = np.concatenate((tabla,dfinita), axis=1)

# Calcula tabla, inicia en columna 3
[n,m] = np.shape(tabla)
diagonal = n-1
j = 3
while (j < m):
    # Añade título para cada columna
    titulo.append('F['+str(j-2)+']')

    # cada fila de columna
    i = 0
    paso = j-2 # inicia en 1
    while (i < diagonal):
        denominador = (xi[i+paso]-xi[i])
        numerador = tabla[i+1,j-1]-tabla[i,j-1]
        tabla[i,j] = numerador/denominador
        i = i+1
    diagonal = diagonal - 1
    j = j+1

# POLINOMIO con diferencias Divididas
# caso: puntos equidistantes en eje x
dDividida = tabla[0,3:]
n = len(dfinita)

# expresión del polinomio con Sympy
x = sym.Symbol('x')
polinomio = fi[0]
print(f"P(0) = {polinomio}")
for j in range(1,n,1):
    factor = dDividida[j-1]
    termino = 1
    
    for k in range(0,j,1):
        termino = termino*(x-xi[k])
        
    Error = termino*factor
    Errorx = sym.lambdify(x,Error)
    print(f"Error P({j-1})  = {Error} ")
    print(f"Error P({j-1})  = {Errorx(Xvalor)}")
    print(' ')
    
    polinomio = polinomio + termino*factor
    print(f"P({j}) = {polinomio}")
    polisimple = polinomio.expand()
    print(f"P({j}) = {polisimple}")
    px = sym.lambdify(x,polisimple)
    print(f"P({j}){px(Xvalor)}")

# polinomio para evaluacion numérica



# SALIDA
np.set_printoptions(precision = 4)
print(' ')
print(' ')
print('Tabla Diferencia Dividida')
print([titulo])
print(tabla)
print(' ')
print(' ')
print('dDividida: ')
print(dDividida)
print(' ')
print(' ')
#print('polinomio: ')
#print(polinomio)
#print('polinomio simplificado: ' )
#print(polisimple)
```

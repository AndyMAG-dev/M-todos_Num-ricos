## Optimizacion

- Es la determinacion de los extremos de una funcion

#### funcion Unimodal
- Cuando tiene 1 solo punto extremo, sea un maximo o un minimo
- Puede ser en todo su domimio o en un intervalo

#### Metodos de busqueda
Sin restricciones: ``(no se conoce el intervalo)``
- Simultanea: Es lo mismo que secuencial, pero no te detienes cuando encuentras el maximo, analizas todo el intervalo y luego buscas
    - Puede ser uniforme o acelerada
- Secuencial:
    - **Uniforme**: Ventajas: Logras la exactitud que quieras, sabes el tamaño del intervalo resultante
    - **Acelerada**: Ecnuentras el maximo mas rapido pero no sabes el tamaño del intervalo resultante

Con restricciones: ``(Si se conoce el intervalo)`` (Se usan para achicar el intervalo una vez se tiene)
- **Biseccion**
- **Regla de Oro**


#### ERRORES
**Estos metodos no tienen errores, el unico error que hay es que si defines el punto maximo o minimo el error que se genera es la mitad del intervalo**

## Sin restricciones

### ``Busqueda Secuencial uniforme``

#### Que te dan:
- ``Y`` = sen(x)    -> La funcion
- ``S`` = 0.1       -> El paso, Aqui es una S
- ``X0`` = 0        -> X sub cero por donde vamos a iniciar

#### Algoritmo
- ``Yi`` es la funcion evaluada en Xi
- ``Xi`` = Xi-1 + S
- Iteras comparando Yi con Yi-1 hasta que el signo cambia

#### Ejemplo

i| Xi|Yi  
-|---|----
0|0.0|0.09
1|0.1|0.10   
2|0,2|0.50
3|0,3|0.42   
4|0.4|0.30    

Y0 < Y1 < Y2 > Y3 > Y4  ->  El signo cambia en Y2
Si el cambio es <> es un maximo
Si el cambio es >< es un minimo

#### Respuesta
- Intervalo de retorno :  **Y1 < X < Y3**
- Intervalo de retorno : X* € [A , B]
- Si te piden el punto maximo o minimo lo que debes dar es **el punto medio del intervalo**, y lo evaluas
- El error que se genera al dar el punto maximo es **la mitad del intervalo**


```python
# Declaracion de la funcion
import math

def f(x):
    return math.exp(x) * math.cos(x) 
```

```python

# declaracion de las variables   
X0 = 0.5
S = 0.1  #Paso

#variables necesarias

i = 0
redondeado = True
iteraciones = 7


```

```python
i = 0
Xi = X0

print("i".ljust(20) + "X".ljust(20) + "Y".ljust(20) + "Comparaciones")

Yi = f(Xi)
if redondeado:
    print(str(i).ljust(15) + "{:.8f}".format(Xi)  + " ".ljust(10) + ( "{:.8f}".format(Yi)) + " ".ljust(10) + ("Inicio") )
else:
    print(str(i).ljust(20) + str(Xi).ljust(20) + str(Yi).ljust(20) + ("Inicio") )


while i <= iteraciones:
        
    Yanterior = Yi
    Xi += S
    i+=1
    Yi = f(Xi)
    
    if redondeado:
        print(str(i).ljust(15) + "{:.8f}".format(Xi)  + " ".ljust(10) + ( "{:.8f}".format(Yi)) + " ".ljust(10) + ("Subiendo" if ((Yi - Yanterior) >= 0) else "Bajando") )
    else:
        print(str(i).ljust(20) + str(Xi).ljust(20) + str(Yi).ljust(20) + ("Subiendo" if ((Yi - Yanterior) >= 0) else "Bajando") )

```

### ``Busqueda Secuencial Acelerada``

Es lo mismo que el uniforme pero

Xi = Xi-1 + S

S = 2*S

```python
i = 0
Xi = X0

print("i".ljust(20) + "X".ljust(20) + "Y".ljust(20) + "Comparaciones")

Yi = f(Xi)
if redondeado:
    print(str(i).ljust(15) + "{:.8f}".format(Xi)  + " ".ljust(10) + ( "{:.8f}".format(Yi)) + " ".ljust(10) + ("Inicio") )
else:
    print(str(i).ljust(20) + str(Xi).ljust(20) + str(Yi).ljust(20) + ("Inicio") )


while i <= iteraciones:
        
    Yanterior = Yi
    Xi += S
    i+=1
    Yi = f(Xi)
    S *= 2
    
    if redondeado:
        print(str(i).ljust(15) + "{:.8f}".format(Xi)  + " ".ljust(10) + ( "{:.8f}".format(Yi)) + " ".ljust(10) + ("Subiendo" if ((Yi - Yanterior) >= 0) else "Bajando") )
    else:
        print(str(i).ljust(20) + str(Xi).ljust(20) + str(Yi).ljust(20) + ("Subiendo" if ((Yi - Yanterior) >= 0) else "Bajando") )

```

## Con restricciones

### Biseccion

#### Que te dan:
- Y = f(x)      -> La funcion
- [a , b]       -> El intervalo
- E             -> El error, tolerancia o condicion de parada
- d (δ delta)   -> es 0.1 * E , obligado o te dan d o te dan delta

#### Calculos necesarios
- L = b-a
- X1 = (a + b)/2 - d/2
- X2 = (a + b)/2 + d/2

#### Algoritmo
- Te dan intervalo, error y funcion
    ciclo
- calculas d, X1, X2, L
- Luego evaluas f(X1) y f(X2) para optener Y1 y Y2
- Si Y1 > Y2 -> a=X1
- Si Y1 < Y2 -> b=X2
- Si ninguno a=X1 y b=X2
- Se repite el ciclo mientras que L sea >= que E

#### Consideraciones
- Si no te dan Error ni delta, el delta se calcula:
    0.1 / L

#### Inciso extra (Iteraciones necesarias para llegar a un intervalo dado)
L_intervalo = L0 / 2^(n/2)

ln(2^(N/2)) = ln(L0/L_intervalo)

(N/2) * ln(2) = ln(L0/L_intervalo)

N = ln(L0/L_intervalo) / ln(2) * 2

N/2 = iteraciones

N = experimentos o Evaluaciones

N debe ser par y si da con coma se aproxima al mayor entero PAR


#### Ejemplo

i| a | b | L | X1 | X2 | Y1 | Y2
-|---|---|---|----|----|----|---
1|3|5|2|3.955|4.005|0.000399|0.000401
2|3|4.005|1.005|3.4975|3.5075|3.0888|2.9841
3|3|4|0.5|3.7|3.7|0.9|0.8


#### Respuesta
- Intervalo de retorno :  Y1 < X < Y3
- Intervalo de retorno : X* € [A , B]
- Si te piden el punto maximo o minimo lo que debes dar es el punto medio del intervalo, y lo evaluas
- El error que se genera al dar el punto maximo es la mitad del intervalo



```python
# Declaracion de la funcion
import math

def funcion(x):
    return math.exp(x) * math.cos(x) 
```

```python
a = 0.7
b = 0.9   #valores del intervalo
E = 0.005   # tolerancia o intervalo final
d = 0.1 * E  # o te lo pueden dar DELTA

i = 0
redondeado = True
L = 99999999999999
L0 = b-a

print("En teoria como buscas un intervalo donde L sea menor que E, la tabla quedaria hasta la ultima L, luego no hay que escribir los X ni Y de esa iteracion")
print("")
print("i".ljust(20) + "a".ljust(20) + "b".ljust(20) + "L".ljust(20) + "X1".ljust(20) + "X2".ljust(20) + "Y1".ljust(20) + "Y2")

while(L >= E):
    
    L = b-a
    X1 = (a + b)/2 - d/2
    X2 = (a + b)/2 + d/2

    Y1 = funcion(X1)
    Y2 = funcion(X2)
    
    if redondeado:
        print(str(i).ljust(15) + "{:.8f}".format(a)  + " ".ljust(10) + ( "{:.8f}".format(b)) + " ".ljust(10) + ( "{:.8f}".format(L)) + " ".ljust(10) + ( "{:.8f}".format(X1)) + " ".ljust(10) + ( "{:.8f}".format(X2))+ " ".ljust(10) + ( "{:.8f}".format(Y1)) + " ".ljust(10) + ( "{:.8f}".format(Y2)))
    else:
        print(str(i).ljust(20) + str(a).ljust(20) + str(b).ljust(20) + str(L).ljust(20) + str(X1).ljust(20) + str(X2).ljust(20) + str(Y1).ljust(20) + str(Y2))
    
    if(Y1>Y2):
        a = X1
    elif(Y1<Y2):
        b = X2
    else:
        a = X1
        b = X2
        
    i += 1
```

```python
print("L_intervalo = L0 / 2^(n/2)")
print("N = ln(L0/L_intervalo) / ln(2) * 2")
print(f"N = {math.log(L0/E, math.e) / math.log(2, math.e) * 2 }")

N = math.log(L0/E, math.e) / math.log(2, math.e) * 2 

print()
print(f"N/2 = iteraciones = {N/2}")
print(f"N = experimentos o evaluaciones = {N}")

print()
print("N debe ser par y si da con coma se aproxima al mayor entero PAR")
```

### Seccion de oro

#### Que te dan:
- Y = f(x)      -> La funcion
- [a , b]       -> El intervalo
- E             -> El error, tolerancia o condicion de parada
- G = 0,618034  -> No se que mierda es, numero de ORO CONSTANTE

#### Calculos necesarios
- L = b-a
- Factor = 1 - G = 0.381966 (En el libro dice que Factor = 0.382 Este no)
- X1 = a + Factor * L
- X2 = b - Factor * L

#### Algoritmo
- Te dan intervalo, error y funcion
    ciclo
- calculas X1, X2, L
- Luego evaluas f(X1) y f(X2) para optener Y1 y Y2
- Si Y1 > Y2 -> a=X1, X1=X2, Y1 = Y2, L=b-a, X2 = b-Factor*L, Y2=F(X2)
- Si Y1 < Y2 -> b=X2, X2=X1, Y2 = Y1, L=b-a, X1 = a+Factor*L, Y1=F(X1)
- Se repite el ciclo mientras que L sea >= que E


#### Inciso extra (Iteraciones necesarias para llegar a un intervalo dado)
L_intervalo/L0 >= G^(n-1)

ln(L_intervalo/L0) >= n-1 * ln(G)

ln(L_intervalo/L0) / ln(G) <= n-1 


N-1 = iteraciones

N = experimentos o evaluaciones

si da con coma se aproxima al mayor entero (No hace falta ser par)


#### Ejemplo

i| a | b | L | X1 | X2 | Y1 | Y2
-|---|---|---|----|----|----|---
1|3|5|2|3.955|4.005|0.000399|0.000401
2|3|4.005|1.005|3.4975|3.5075|3.0888|2.9841
3|3|4|0.5|3.7|3.7|0.9|0.8


#### Respuesta
- Intervalo de retorno :  Y1 < X < Y3
- Intervalo de retorno : X* € [A , B]
- Si te piden el punto maximo o minimo lo que debes dar es el punto medio del intervalo, y lo evaluas
- El error que se genera al dar el punto maximo es la mitad del intervalo

```python
a = 3
b = 5   #valores del intervalo
E = 0.1   # tolerancia o intervalo final
G = 0.618034
Factor = 0.381966

i = 0
redondeado = True
L = 99999999999999

print("En teoria como buscas un intervalo donde L sea menor que E, la tabla quedaria hasta la ultima L, luego no hay que escribir los X ni Y de esa iteracion")
print("")
print("i".ljust(20) + "a".ljust(20) + "b".ljust(20) + "L".ljust(20) + "X1".ljust(20) + "X2".ljust(20) + "Y1".ljust(20) + "Y2")

L0 = b-a
L = b-a
X1 = a + Factor * L
X2 = b - Factor * L
Y1 = funcion(X1)
Y2 = funcion(X2)

if redondeado:
    print(str(i).ljust(15) + "{:.8f}".format(a)  + " ".ljust(10) + ( "{:.8f}".format(b)) + " ".ljust(10) + ( "{:.8f}".format(L)) + " ".ljust(10) + ( "{:.8f}".format(X1)) + " ".ljust(10) + ( "{:.8f}".format(X2))+ " ".ljust(10) + ( "{:.8f}".format(Y1)) + " ".ljust(10) + ( "{:.8f}".format(Y2)))
else:
        print(str(i).ljust(20) + str(a).ljust(20) + str(b).ljust(20) + str(L).ljust(20) + str(X1).ljust(20) + str(X2).ljust(20) + str(Y1).ljust(20) + str(Y2))

while(L >= E):
    
    if(Y1>Y2):
        a = X1
        X1 = X2
        Y1 = Y2
        L = b-a
        X2 = b-Factor*L
        Y2 = funcion(X2)
        
    elif(Y1<Y2):
        b = X2
        X2 = X1
        Y2 = Y1
        L = b-a
        X1 = a + Factor*L
        Y1 = funcion(X1)
    
        
    i += 1
    if redondeado:
        print(str(i).ljust(15) + "{:.8f}".format(a)  + " ".ljust(10) + ( "{:.8f}".format(b)) + " ".ljust(10) + ( "{:.8f}".format(L)) + " ".ljust(10) + ( "{:.8f}".format(X1)) + " ".ljust(10) + ( "{:.8f}".format(X2))+ " ".ljust(10) + ( "{:.8f}".format(Y1)) + " ".ljust(10) + ( "{:.8f}".format(Y2)))
    else:
        print(str(i).ljust(20) + str(a).ljust(20) + str(b).ljust(20) + str(L).ljust(20) + str(X1).ljust(20) + str(X2).ljust(20) + str(Y1).ljust(20) + str(Y2))
        
```

```python
print("L_intervalo / L0  >=  G^(n-1)")
print("N-1 >= ln(L_intervalo/L0) / ln(G)")
print(f"N = {math.log(E/L0, math.e) / math.log(G, math.e) + 1}")

N = math.log(E/L0, math.e) / math.log(G, math.e) + 1

print()
print(f"N-1 = iteraciones = {N-1}")
print(f"N = experimentos o evaluaciones = {N}")

print()
print("si da con coma se aproxima al mayor entero, No hace falta ser par")
```

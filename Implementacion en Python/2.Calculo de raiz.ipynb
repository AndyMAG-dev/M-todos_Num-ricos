### Lista Completa de Funciones en el Módulo math

Aquí tienes una lista con algunas de las funciones más útiles en el módulo math:

- Constantes:
  - `math.e` - La constante e (base del logaritmo natural)
  - `math.pi` - La constante π (pi)

- Funciones Aritméticas:
  - `math.ceil(x)` - Redondea hacia arriba al entero más cercano
  - `math.floor(x)` - Redondea hacia abajo al entero más cercano

- Funciones Exponenciales y Logarítmicas:
  - `math.exp(x)` - e elevado a la x (exponencial)

  - `math.log10(x)` - Logaritmo en base 10

  - `math.sin(x)` - Seno de x (en radianes)

  - `math.tan(x)` - Tangente de x (en radianes)
  - `math.asin(x)` - Arcoseno de x (resultado en radianes)
  - `math.acos(x)` - Arcocoseno de x (resultado en radianes)
  - `math.atan(x)` - Arcotangente de x (resultado en radianes)
  - `math.atan2(y, x)` - Arcotangente de y / x (resultado en radianes)

- Funciones Hiperbólicas:
  - `math.sinh(x)` - Seno hiperbólico de x
  - `math.cosh(x)` - Coseno hiperbólico de x
  - `math.tanh(x)` - Tangente hiperbólica de x
  - `math.asinh(x)` - Arcoseno hiperbólico de x
  - `math.acosh(x)` - Arcocoseno hiperbólico de x
  - `math.atanh(x)` - Arcotangente hiperbólica de x

- Funciones Angulares:
  - `math.degrees(x)` - Conv

### Metodo de biseccion
Sea la ecuación f(x) = 0 y un intervalo [a, b] tales que:
1.	f(x) es continua en [a, b].   (Graficando)
2.	 f(x) posee signos diferentes en a y en b, o sea, f(a) · f(b) < 0.   (Funcion ARRIBA) y (Graficando)
3.	En el intervalo la ecuación tiene raíz única.     (Graficando)

### Metodo de Newton Raphson

Sea la ecuación f(x)=0  y un intervalo [a, b]  tales que:
1. f(x) es continua en [a, b].  (graficar)
2. f(x) posee signos diferentes en a y en b,  o sea, f(a)f(b)<0. (funcion arriba)
3. En el intervalo la ecuación tiene raíz única. (graficar)
4. f´(x) y  f´´(x) son continuas y no nulas en [a, b]. (funcion mas graficar)
5. x_0  ϵ [a, b] : f(x_0 )f´´(x_0)>0

Consideraciones mias:
- se necesita un Xo que es un valor decidido por mi dentro del intervalo
- el valor tiene que cumplir con la ecuacion f(x) * df2(x) > 0. osea que si es concava hacia arriba el valor la funcion evualuada en X debe ser positivo, y si es concava hacia abajo debe ser negativo




```python
import sympy
from math import *

from decimal import Decimal, getcontext

getcontext().prec = 28

def NewtonR(Xo, tol, iter, redondeado):
    x = sympy.symbols("x")
    f = " x * exp(x) -2"                           ### modificar aqui la funcion
    df = sympy.diff(f)
    fl = sympy.lambdify(x,f)
    dfl = sympy.lambdify(x,df)
    i = 0
    error = 1.00
    
    print(f"f(x) = {f}")
    print(f"df(x) = {df}")
    print(" ")
    
    print("i".ljust(20) + "Xi".ljust(20) + "Xi+1".ljust(20) + "error")
    
    while i < iter and error > tol:
        X1 = Xo - fl(Xo)/dfl(Xo)
        error = Decimal(abs(X1 -Xo))
        i += 1
        
        if redondeado:
            print(str(i).ljust(15) + "{:.8f}".format(Xo) + " ".ljust(10) +   "{:.8f}".format(X1) +" ".ljust(15) +  ( "{:.8f}".format(error)) )
        else:
            print(str(i).ljust(20) + str(Xo).ljust(20) + str(X1).ljust(20) + str(error).ljust(20) )
        
        
        Xo = X1
        
valor = 1
tol = 0.000000001
iter = 3
redondeo = False
NewtonR(valor, tol ,iter, redondeo)                   ### MODIFICAR AQUI LOS PARAMETROS
```
